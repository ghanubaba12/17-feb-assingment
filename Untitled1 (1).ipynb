```python
Q4. Using the database and the collection created in question number 3, write a code to insert one record,
and insert many records. Use the find() and find_one() methods to print the inserted record.
ans-Sure! Here's an example code that inserts one record and multiple records into the customers collection and then uses the find() and find_one() methods to print the inserted records:

python
Copy code
# Importing the required libraries
from pymongo import MongoClient

# Creating a connection to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Accessing the "mydatabase" database and "customers" collection
db = client['mydatabase']
collection = db['customers']

# Inserting one record into the "customers" collection
customer1 = {'name': 'Alice', 'email': 'alice@example.com', 'phone': '111-111-1111'}
inserted_record = collection.insert_one(customer1)
print('Inserted record ID:', inserted_record.inserted_id)

# Inserting many records into the "customers" collection
customers = [
    {'name': 'Bob', 'email': 'bob@example.com', 'phone': '222-222-2222'},
    {'name': 'Charlie', 'email': 'charlie@example.com', 'phone': '333-333-3333'},
    {'name': 'David', 'email': 'david@example.com', 'phone': '444-444-4444'}
]
inserted_records = collection.insert_many(customers)
print('Inserted record IDs:', inserted_records.inserted_ids)

# Retrieving the inserted record using the find_one() method
retrieved_record = collection.find_one({'_id': inserted_record.inserted_id})
print('Retrieved record:', retrieved_record)

# Retrieving all inserted records using the find() method
retrieved_records = collection.find()
print('Retrieved records:')
for record in retrieved_records:
    print(record)
In this code, we first create a connection to MongoDB and access the mydatabase database and the customers collection.

We then insert one record into the customers collection using the insert_one() method and print the ID of the inserted record.

Next, we insert multiple records into the customers collection using the insert_many() method and print the IDs of the inserted records.

We then use the find_one() method to retrieve the record with the ID of the first inserted record and print it to the console.

Finally, we use the find() method to retrieve all inserted records and print them to the console using a for loop.




```

```python
Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to
demonstrate this.
ans-The find() method in MongoDB is used to retrieve documents from a collection. You can pass a query object to the find() method to filter the documents that are returned.

Here's an example code that demonstrates how to use the find() method to query the customers collection in the mydatabase database for all documents with the name "John":

python
Copy code
# Importing the required libraries
from pymongo import MongoClient

# Creating a connection to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Accessing the "mydatabase" database and "customers" collection
db = client['mydatabase']
collection = db['customers']

# Querying the "customers" collection for all documents with the name "John"
query = {'name': 'John'}
results = collection.find(query)

# Printing the results
for result in results:
    print(result)
In this code, we first create a connection to MongoDB and access the mydatabase database and the customers collection.

We then create a query object with the key "name" and the value "John", and pass it to the find() method. This filters the documents in the customers collection to only include those with the name "John".

Finally, we print the results of the query to the console using a for loop. This will print all documents in the customers collection with the name "John".




```

```python
Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
ans-
The sort() method in MongoDB is used to sort the documents in a collection based on one or more fields. It takes a dictionary of field and order pairs, where the field is the name of the field to sort by, and the order is either 1 for ascending or -1 for descending.

Here's an example code that demonstrates how to use the sort() method to sort the customers collection in the mydatabase database by the name field in ascending order:

python
Copy code
# Importing the required libraries
from pymongo import MongoClient

# Creating a connection to MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Accessing the "mydatabase" database and "customers" collection
db = client['mydatabase']
collection = db['customers']

# Sorting the "customers" collection by the "name" field in ascending order
sort_order = [('name', 1)]
results = collection.find().sort(sort_order)

# Printing the sorted results
for result in results:
    print(result)
In this code, we first create a connection to MongoDB and access the mydatabase database and the customers collection.

We then create a sort order object with the key "name" and the value 1 (for ascending order), and pass it to the sort() method. This sorts the documents in the customers collection by the name field in ascending order.

Finally, we print the sorted results to the console using a for loop. This will print all documents in the customers collection, sorted by the name field in ascending order.




```

```python
Q7.Explain why delete_one(), delete_many(), and drop() is used.
ans-
delete_one(), delete_many(), and drop() are all methods in MongoDB that are used for removing data from a collection.

delete_one(filter) method is used to delete a single document that matches the specified filter criteria. If multiple documents match the filter, only the first document will be deleted.

delete_many(filter) method is used to delete all documents that match the specified filter criteria. This method can be used to delete multiple documents at once.

drop() method is used to completely remove a collection from the database. This method is irreversible and will permanently delete all documents in the collection.

These methods are used when data is no longer needed or when it is necessary to make significant changes to a database's structure. For example, delete_one() and delete_many() can be used to remove outdated or irrelevant data from a collection, while drop() can be used to delete an entire collection that is no longer needed.

It is important to use these methods with caution as data loss can occur if not used properly. It is recommended to backup the database before making any major changes.





```
